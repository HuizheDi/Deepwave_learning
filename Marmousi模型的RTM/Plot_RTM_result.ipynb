```python
import torch
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
```

```python
device = 'cpu'
```

```python
observed_data = (
    torch.from_file('Marmousi_RTM_no_smooth_model.bin',
                    size=1151*376)
    .reshape(1151, 376)
)
```

```python
observed_data.detach()
```

```python
# Plot
vmin, vmax = torch.quantile(observed_data.detach(),
                            torch.tensor([0.05, 0.95]).to(device))
plt.figure(figsize=(10.5, 3.5))
plt.imshow(observed_data.detach().cpu().T, aspect='auto', cmap='gray',vmin=vmin, vmax=vmax)
plt.title('Marmousi RTM with no smooth velocity model')
plt.savefig('Marmousi_RTM_no_smooth_model.jpg')
```
